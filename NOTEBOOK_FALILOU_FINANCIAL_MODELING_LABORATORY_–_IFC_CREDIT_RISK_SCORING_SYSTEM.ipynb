# 2. NOTEBOOK

## 2.1 Preprocessing

```python


# Gestion valeurs manquantes

def impute_group(df, col, group_cols, method="median"):
    if method == "median":
        return df.groupby(group_cols)[col].transform(lambda x: x.fillna(x.median()))
    elif method == "mean":
        return df.groupby(group_cols)[col].transform(lambda x: x.fillna(x.mean()))

num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    df[col] = impute_group(df, col, ["pays_implantation", "secteur_activite"])


# Encodage catégoriel

from sklearn.preprocessing import OneHotEncoder

cat_cols = [
    "type_client","secteur_activite","taille_entreprise",
    "statut_juridique","pays_implantation","region",
    "ville","zone_urbaine_rurale","type_intervention_ifc",
    "categorie_esg_ifc"
]

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded = encoder.fit_transform(df[cat_cols])

encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
df = pd.concat([df.drop(columns=cat_cols), encoded_df], axis=1)


# Normalisation

from sklearn.preprocessing import RobustScaler

num_cols = df.select_dtypes(include=np.number).columns
scaler = RobustScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])


# Correction biais multi-pays

def normalize_country(df, col):
    return df.groupby("score_risque_pays")[col].transform(lambda x: (x - x.mean()) / (x.std()+1e-6))

for col in num_cols:
    df[col] = normalize_country(df, col)
```

## 2.2 Feature enginneering

```python

# Risque financier

df["ratio_dette_ca"] = df["encours_dette"] / (df["chiffre_affaires"] + 1)
df["ratio_endettement_calc"] = df["encours_dette"] / (df["total_actifs"] + 1)
df["dscr_calc"] = df["ebitda"] / (df["service_dette"] + 1)


# Interaction macro

df["inflation_x_dette"] = df["inflation"] * df["encours_dette"]
df["croissance_x_secteur"] = df["croissance_pays"] * df["risque_sectoriel"]


# Score ESG

df["score_env"] = (
    df["impact_environnemental"] +
    df["gestion_risques_env"] +
    df["plan_reduction_carbone"]
) / 3

df["score_social"] = (
    df["impact_social"] +
    df["inclusion_genre"] +
    df["taux_femmes_cadres"]
) / 3

df["score_gouv"] = (
    df["qualite_management"] +
    df["transparence_financiere"] +
    (1 - df["risque_corruption"])
) / 3

df["score_esg"] = (
    df["score_env"] * 0.3 +
    df["score_social"] * 0.3 +
    df["score_gouv"] * 0.4
)


# Impact développement

df["impact_dev"] = df["impact_social"] + df["inclusion_genre"]


# Score IFC global

df["score_ifc"] = (
    df["score_credit"] * 0.4 +
    df["score_esg"] * 0.2 +
    df["impact_dev"] * 0.2 +
    df["score_risque_pays"] * 0.2
)
```

## 2.3 Benchmark Global

```python
pip install tabnet
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
import matplotlib.pyplot as plt




# NETTOYAGE GLOBAL


X_clean = X.select_dtypes(include=['number'])

if "classe_risque" in X_clean.columns:
    X_clean = X_clean.drop(columns=["classe_risque"])

mask = ~y.isna()
X_clean = X_clean[mask]
y_clean = y[mask]

X_clean = X_clean.fillna(0)



# FEATURE GRAPH (approximation Graph ML)


X_clean["mean_risk_country"] = X_clean.groupby("score_risque_pays")["score_risque_pays"].transform("mean")
X_clean["interaction_sector_macro"] = X_clean["croissance_pays"] * X_clean["risque_sectoriel"]



# MODELES


models = {
    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42),
    "XGBoost": xgb.XGBRegressor(n_estimators=200, random_state=42),
    "LightGBM": lgb.LGBMRegressor(n_estimators=200, random_state=42),
    "CatBoost": CatBoostRegressor(iterations=200, verbose=0, random_state=42),
    "MLP": MLPRegressor(hidden_layer_sizes=(128,64), max_iter=300, random_state=42),
}



# SOTA

try:
    from pytorch_tabnet.tab_model import TabNetRegressor
    models["TabNet"] = TabNetRegressor(verbose=0)
except:
    print("TabNet non installé")

try:
    from tabpfn import TabPFNRegressor
    models["TabPFN"] = TabPFNRegressor()
except:
    print("TabPFN non installé")

models["FT-Transformer (proxy)"] = MLPRegressor(hidden_layer_sizes=(256,128,64), max_iter=300)



# KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
results = {}



# TRAIN + CROSS VALIDATION

for name, model in models.items():
    print(f"Training {name}...")
    scores = []

    for train_idx, val_idx in kf.split(X_clean):
        X_tr, X_val = X_clean.iloc[train_idx], X_clean.iloc[val_idx]
        y_tr, y_val = y_clean.iloc[train_idx], y_clean.iloc[val_idx]

        try:
            model.fit(X_tr, y_tr)
            preds = model.predict(X_val)
            scores.append(mean_squared_error(y_val, preds))
        except Exception as e:
            print(f"Erreur avec {name} :", e)
            scores.append(np.nan)

    results[name] = np.nanmean(scores)



# RESULTATS

results_df = pd.DataFrame.from_dict(results, orient='index', columns=['MSE'])
results_df = results_df.sort_values(by='MSE')

print("\nRésultats Benchmark :")
print(results_df)


# MEILLEUR MODELE

best_model_name = results_df.index[0]
print(f"\nMeilleur modèle : {best_model_name}")



# GRAPHIQUE AMÉLIORÉ

plt.figure(figsize=(10,6))

colors = plt.cm.tab20.colors[:len(results_df)]

bars = plt.bar(results_df.index, results_df["MSE"], color=colors)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height:.4f}",
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.xticks(rotation=45)
plt.title("FF_Modeling_Lab - Comparaison des modèles (MSE: Mean Squared Error)", fontsize=14)
plt.xlabel("Modèles")
plt.ylabel("Erreur MSE")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
